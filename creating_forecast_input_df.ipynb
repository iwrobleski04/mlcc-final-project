```python
#dropping default values... 
to_drop = df2[df2['NIR'] == -999]
to_drop.shape
print(df2.shape)
df2 = df2[(df2['green'] != -999) & (df2['blue'] != -999) & (df2['SWIR'] != -999) & (df2['NIR'] != -999)]
print(df2.shape)

```

```python
#set a cloud coverage limit (drop anything with cloud coverage over a chosen barrier)
#create multiple datasets with different cloud coverages 
#20% 
df_under20 = df2[df2['cloud_score'] <= 0.2]
print("Under 20%: ", df_under20.shape)
#30%
df_under30 = df2[df2['cloud_score'] <= 0.3]
print("Under 30%: ", df_under30.shape)
#50% 
df_under50 = df2[df2['cloud_score'] <= 0.5]
print("Under 50%: ", df_under50.shape)
#70%
df_under70 = df2[df2['cloud_score'] <= 0.7]
print("Under 70%: ", df_under70.shape)
```

```python
#removing duplicates ... keep the version that has a lower cloud score
df_under20_clean = (df_under20.sort_values('cloud_score', ascending = True).drop_duplicates(subset = ['cyanobacteria_abundance','sample_date', 'satellite_date']))
df_under30_clean = (df_under30.sort_values('cloud_score', ascending = True).drop_duplicates(subset = ['cyanobacteria_abundance','sample_date', 'satellite_date']))
df_under50_clean = (df_under50.sort_values('cloud_score', ascending = True).drop_duplicates(subset = ['cyanobacteria_abundance','sample_date', 'satellite_date']))
df_under70_clean = (df_under70.sort_values('cloud_score', ascending = True).drop_duplicates(subset = ['cyanobacteria_abundance','sample_date', 'satellite_date']))

```

```python
#display number of unique samples + the number of samples w/ n previous...
#ie how many with 9, 8, 7 --> used to determine final sample size + shape of input
a_list = [df_under20_clean, df_under30_clean, df_under50_clean, df_under70_clean]
for temp in a_list:
    counts = (
        temp.groupby(['sample_date', 'latitude', 'longitude', 'cyanobacteria_abundance'])
          .size()
          .reset_index(name='n_obs')
    )

    n_7 = (counts['n_obs'] >= 7).sum()
    n_8 = (counts['n_obs'] >= 8).sum()
    n_9 = (counts['n_obs'] >= 9).sum()

    print(n_7, n_8, n_9)

    print(n_7 / 6327 , n_8 / 6327 , n_9 / 6327)

```

```python
#returning 3 possible training data frames for the forecaster

#1 - 7 time points before w/ a cloud coverage limit @ 0.3
df_one = (
    df_under30_clean.sort_values('satellite_date', ascending=False)
      .groupby(['cyanobacteria_abundance', 'sample_date'])
        .filter(lambda x: len(x) >= 7)
        .groupby(['cyanobacteria_abundance', 'sample_date'])
        .head(7)
)

#2 - 9 time points w/ a cloud coverage limit @ 0.7
df_two = (
    df_under70_clean.sort_values('satellite_date', ascending=False)
      .groupby(['cyanobacteria_abundance', 'sample_date'])
        .filter(lambda x: len(x) >= 9)
        .groupby(['cyanobacteria_abundance', 'sample_date'])
        .head(9)
)
#3 - 8 time points w/ a cloud coverage limit @ 0.5
df_three = (
    df_under50_clean.sort_values('satellite_date', ascending=False)
      .groupby(['cyanobacteria_abundance', 'sample_date'])
        .filter(lambda x: len(x) >= 8)
        .groupby(['cyanobacteria_abundance', 'sample_date'])
        .head(8)
)

print(df_one.shape)
print(df_two.shape)
print(df_three.shape)
print(df_one.shape[0] / 7)
print(df_two.shape[0] / 9)
print(df_three.shape[0] / 8)
```

```python
#now calculated columns
#put in datetime
df_one['sample_date'] = pd.to_datetime(df_one['sample_date'])
df_one['satellite_date'] = pd.to_datetime(df_one['satellite_date'])
#sorting
df_one = df_one.sort_values(['sample_date', 'cyanobacteria_abundance'])

#calculating statistics
group_cols = ['sample_date', 'cyanobacteria_abundance']

def compute_time_features(group):
    group = group.sort_values('satellite_date')
    
    #create list of differences within each sample
    diffs = group['satellite_date'].diff().dt.days.dropna()
    
    #calculating the total difference
    tot_diff = (group['sample_date'].iloc[0] - group['satellite_date'].iloc[0]).days
    
    return pd.Series({
        'tot_diff_days': tot_diff,
        'mean_diff_days': diffs.mean(),
        'median_diff_days': diffs.median(),
        'std_diff_days': diffs.std()
    })

time_features_one = df_one.groupby(group_cols).apply(compute_time_features).reset_index()

print(time_features_one.shape)
time_features_one.head(5)
```

```python
#now do the same for dfs 2 and 3
#dataframe 2
#put in datetime
df_two['sample_date'] = pd.to_datetime(df_two['sample_date'])
df_two['satellite_date'] = pd.to_datetime(df_two['satellite_date'])
#sorting
df_two = df_two.sort_values(['sample_date', 'cyanobacteria_abundance'])

time_features_two = df_two.groupby(group_cols).apply(compute_time_features).reset_index()

print(time_features_two.shape)
time_features_two.head(5)

#dataframe 3: 
#put in datetime
df_three['sample_date'] = pd.to_datetime(df_three['sample_date'])
df_three['satellite_date'] = pd.to_datetime(df_three['satellite_date'])
#sorting
df_three = df_three.sort_values(['sample_date', 'cyanobacteria_abundance'])

time_features_three = df_three.groupby(group_cols).apply(compute_time_features).reset_index()

print(time_features_three.shape)
time_features_three.head(5)
```

```python
#now merge everything back to final dataframes
df_one_final = df_one.merge(time_features_one, on=group_cols)
print(df_one_final.shape)

df_two_final = df_two.merge(time_features_two, on = group_cols)
print(df_two_final.shape)

df_three_final = df_three.merge(time_features_three, on = group_cols)
print(df_three_final.shape)
```
